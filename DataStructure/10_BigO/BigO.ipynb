# Big O Notation

From simple array operations to complex sorting algorithms, understanding the Big O Notation is critical for building high-performance software solutions. 
 
### 1 - O(1) 
This is the constant time notation. The runtime remains steady regardless of input size. For example, accessing an element in an array by index and inserting/deleting an element in a hash table. 
 
### 2 - O(n) 
Linear time notation. The runtime grows in direct proportion to the input size. For example, finding the max or min element in an unsorted array. 
 
### 3 - O(log n) 
Logarithmic time notation. The runtime increases slowly as the input grows. For example, a binary search on a sorted array and operations on balanced binary search trees. 
 
### 4 - O(n^2) 
Quadratic time notation. The runtime grows exponentially with input size. For example, simple sorting algorithms like bubble sort, insertion sort, and selection sort. 
 
### 5 - O(n^3) 
Cubic time notation. The runtime escalates rapidly as the input size increases. For example, multiplying two dense matrices using the naive algorithm. 
 
### 6 - O(n logn) 
Linearithmic time notation. This is a blend of linear and logarithmic growth. For example, efficient sorting algorithms like merge sort, quick sort, and heap sort 
 
### 7 - O(2^n) 
Exponential time notation. The runtime doubles with each new input element. For example, recursive algorithms solve problems by dividing them into multiple subproblems. 
 
### 8 - O(n!) 
Factorial time notation. Runtime skyrockets with input size. For example, permutation-generation problems. 
 
### 9 - O(sqrt(n)) 
Square root time notation. Runtime increases relative to the input’s square root. For example, searching within a range such as the Sieve of Eratosthenes for finding all primes up to n.

```python
import time
import random

# 1. O(1) - Constant time
def constant_time_operation(arr):
    return arr[0]  # Accessing the first element of the array

# 2. O(n) - Linear time
def linear_time_operation(arr):
    maximum = arr[0]
    for i in arr:
        if i > maximum:
            maximum = i
    return maximum

# 3. O(log n) - Logarithmic time
def logarithmic_time_operation(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# 4. O(n^2) - Quadratic time
def quadratic_time_operation(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]  # Bubble sort style swap

# 5. O(n^3) - Cubic time (Matrix multiplication)
def cubic_time_operation(A, B):
    result = [[0] * len(A) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(A)):
            for k in range(len(A)):
                result[i][j] += A[i][k] * B[k][j]
    return result

# 6. O(n log n) - Linearithmic time (Merge Sort)
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

def merge(left, right):
    sorted_list = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1
    sorted_list.extend(left[i:])
    sorted_list.extend(right[j:])
    return sorted_list

# 7. O(2^n) - Exponential time (Fibonacci recursion)
def exponential_time_fibonacci(n):
    if n <= 1:
        return n
    return exponential_time_fibonacci(n - 1) + exponential_time_fibonacci(n - 2)

# 8. O(n!) - Factorial time (Generating permutations)
from itertools import permutations
def factorial_time_operation(arr):
    return list(permutations(arr))

# 9. O(sqrt(n)) - Square root time (Finding prime numbers using the Sieve of Eratosthenes)
def sqrt_time_operation(n):
    primes = [True] * (n + 1)
    p = 2
    while (p * p <= n):
        if primes[p]:
            for i in range(p * p, n + 1, p):
                primes[i] = False
        p += 1
    prime_numbers = [p for p in range(2, n + 1) if primes[p]]
    return prime_numbers

# Helper function to measure execution time
def measure_time(func, *args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    print(f"Time taken: {end_time - start_time:.6f} seconds")
    return result

# Example of use cases:

# 1. O(1) Example
arr = [random.randint(0, 1000) for _ in range(1000)]
print("O(1):", measure_time(constant_time_operation, arr))

# 2. O(n) Example
print("O(n):", measure_time(linear_time_operation, arr))

# 3. O(log n) Example
sorted_arr = sorted(arr)
print("O(log n):", measure_time(logarithmic_time_operation, sorted_arr, 500))

# 4. O(n^2) Example
print("O(n^2):", measure_time(quadratic_time_operation, arr[:]))

# 5. O(n^3) Example
A = [[random.randint(0, 10) for _ in range(5)] for _ in range(5)]
B = [[random.randint(0, 10) for _ in range(5)] for _ in range(5)]
print("O(n^3):", measure_time(cubic_time_operation, A, B))

# 6. O(n log n) Example
print("O(n log n):", measure_time(merge_sort, arr[:]))

# 7. O(2^n) Example
print("O(2^n):", measure_time(exponential_time_fibonacci, 20))

# 8. O(n!) Example
small_arr = [1, 2, 3, 4]
print("O(n!):", measure_time(factorial_time_operation, small_arr))

# 9. O(sqrt(n)) Example
print("O(sqrt(n)):", measure_time(sqrt_time_operation, 1000))
```


### Key points demonstrated in the script:
1. **O(1)**: Demonstrates constant time operation by accessing the first element of an array.
2. **O(n)**: Linear time is shown by finding the maximum element in an array.
3. **O(log n)**: A binary search implementation is used to illustrate logarithmic time.
4. **O(n²)**: Bubble sort, a simple quadratic time algorithm.
5. **O(n³)**: Naive matrix multiplication is an example of cubic time complexity.
6. **O(n log n)**: Merge sort is used to demonstrate linearithmic time complexity.
7. **O(2^n)**: The Fibonacci sequence with recursion is shown as an example of exponential time.
8. **O(n!)**: Generating all permutations of an array demonstrates factorial time.
9. **O(√n)**: The Sieve of Eratosthenes is used to find primes as an example of square root time complexity.


```python
import matplotlib.pyplot as plt
import numpy as np

# Functions for different Big O complexities

# O(1) - Constant time
def constant_time(n):
    return 1

# O(n) - Linear time
def linear_time(n):
    return n

# O(log n) - Logarithmic time
def logarithmic_time(n):
    return np.log2(n)

# O(n^2) - Quadratic time
def quadratic_time(n):
    return n ** 2

# O(n^3) - Cubic time
def cubic_time(n):
    return n ** 3

# O(n log n) - Linearithmic time
def linearithmic_time(n):
    return n * np.log2(n)

# O(2^n) - Exponential time
def exponential_time(n):
    return 2 ** n

# O(sqrt(n)) - Square root time
def sqrt_time(n):
    return np.sqrt(n)

# Creating input sizes for visualization
n_values = np.linspace(1, 100, 100)

# Adjusting input sizes for exponential complexity
n_values_exponential = np.linspace(1, 20, 20)

# Chart for different Big O complexities
plt.figure(figsize=(12, 8))

# Plotting each Big O complexity
plt.plot(n_values, [constant_time(n) for n in n_values], label='O(1) - Constant')
plt.plot(n_values, [linear_time(n) for n in n_values], label='O(n) - Linear')
plt.plot(n_values, [logarithmic_time(n) for n in n_values], label='O(log n) - Logarithmic')
plt.plot(n_values, [quadratic_time(n) for n in n_values], label='O(n^2) - Quadratic')
plt.plot(n_values, [cubic_time(n) for n in n_values], label='O(n^3) - Cubic')
plt.plot(n_values, [linearithmic_time(n) for n in n_values], label='O(n log n) - Linearithmic')
plt.plot(n_values_exponential, [exponential_time(n) for n in n_values_exponential], label='O(2^n) - Exponential', color='purple')
plt.plot(n_values, [sqrt_time(n) for n in n_values], label='O(sqrt n) - Square root')

# Labeling the chart
plt.yscale('log')  # Using logarithmic scale to better display growth
plt.xlabel('Input size (n)')
plt.ylabel('Time (log scale)')
plt.title('Comparison of Big O Time Complexities')
plt.legend()
plt.grid(True)

# Displaying the chart
plt.show()
```

### Breakdown of the code:
- **Functions:** Each function represents a Big O complexity (O(1), O(n), O(log n), etc.), and they return values based on the input size `n`.
- **Input Data:** The `n_values` array is created using `np.linspace(1, 100, 100)` to represent a range of input sizes from 1 to 100, except for the exponential complexity which uses a smaller range (`n_values_exponential = np.linspace(1, 20, 20)`).
- **Chart Creation:** The `matplotlib.pyplot` library is used to plot the time complexity curves for each function, and the y-axis is scaled logarithmically to better display the growth of the different complexities.
- **Display:** The chart is displayed with legends, gridlines, and labels for clarity. 

This code can be run in any Python environment that supports `matplotlib` and `numpy` to visualize the time complexities.
